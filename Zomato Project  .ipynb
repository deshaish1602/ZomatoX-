 # ZOMATOX INSIGHTS 

## Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Create the Dataframe

```python
df = pd.read_csv("zomato data .csv")

print(df)
print(df.head())
print(df.info())
```

```python
def handleRate(value):
    value = str(value).split('/')
    return float(value[0])

df['rate'] = df['rate'].apply(handleRate)
 

```

## Summary of the Dataframe

```python
df.info()
```

## Type of Restaurant  

```python
sns.countplot(x=df['listed_in(type)'])
plt.xlabel("Type of restaurant")
```

### Conclusion : The majority of the restaurants fall into the dinning category .

### Dining restaurants are preferred by a larger number of individuals.

```python
grouped_data = df.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes' : grouped_data})
plt.plot(result, c="green", marker="o")
plt.xlabel("Type of restaurant",c = "blue" ,size=20)
plt.ylabel("Votes", c="blue", size=20)
 
```

### The majority of restaurants received ratings .

```python
plt.hist(df['rate'], bins=5)
plt.title("Ratings Distribution")
plt.show()
```

###  conclusion : The majority of restaurants received ratings ranging from 3.5 to 4.

### The majority  of couples prefer restaurants with an approximate cost of 300 rupees.

```python
couple_data=df['approx_cost(for two people)']
sns.countplot(x=couple_data)
 
```

### whether online orders receive higher ratings than offline orders .

```python
plt.figure(figsize = (6,6))
sns.boxplot(x = 'online_order', y = 'rate', data = df)
```

### Conclusion : Offline orders received lower ratings in comparison to online orders , which obtained excellent ratings. 

```python
pivot_table = df.pivot_table(index='listed_in(type)', columns='online_order', aggfunc='size')
sns.heatmap(pivot_table, annot=True, cmap='YlGnBu', fmt='d')
plt.title("Heatmap")
plt.xlabel("Online Order")
plt.ylabel("Listed In (Type)")
plt.show()
```

### Conclusion : Dining restaurants primarily accept offline orders, whereas cafes primarily receive online orders. This suggests that clients prefer to place orders in person at restaurnts, but prefer online ordering at cafes.
